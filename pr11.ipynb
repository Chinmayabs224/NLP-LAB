                <li>Corpus: ["The Cat in the Hat", "The quick brown fox jumps over the lazy dog", "The cat jumped over the fence", "A brown dog jumped over the cat"]</li>

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "The Cat in the Hat",
    "The quick brown fox jumps over the lazy dog",
    "The cat jumped over the fence",
    "A brown dog jumped over the cat"
]

tfidf = TfidfVectorizer()
scores = tfidf.fit_transform(corpus)

# Print the TF-IDF scores
features = tfidf.get_feature_names_out()
for doc, vector in zip(corpus, scores.toarray()):
    print(f"Document: {doc}")
    print(dict(zip(features, vector)))
    print()
```
