```python
import nltk
from nltk import pos_tag, word_tokenize
from nltk.tree import Tree

nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

text = "Text processing with POS tagging is interesting."
tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

print("POS Tags:", pos_tags)

# Drawing parse tree (for demonstration)
grammar = "NP: {<DT>?<JJ>*<NN>}"
chunk_parser = nltk.RegexpParser(grammar)
tree = chunk_parser.parse(pos_tags)
tree.draw()
```

```python
import matplotlib.pyplot as plt
from nltk.probability import FreqDist

text = "Text analysis often involves frequent word distributions. Text processing is key."
tokens = word_tokenize(text.lower())
fdist = FreqDist(tokens)

fdist.plot(10, title="Top 10 Most Frequent Words")
```
